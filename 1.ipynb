**Importing all dependencies**

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

%matplotlib inline
```

**Loading Dataset**

```python
data = pd.read_csv('datasets.csv',encoding_errors='ignore')
```

**Data Cleaning**

```python
data.isnull().sum()

#Dropping all missing values rows 
data.dropna(inplace=True)

data.isnull().sum()
```

```python
data.shape
```

```python
# For duplicate records 
data.duplicated().sum()

#deleting all duplicated rows 
data[data.duplicated()]

data.drop_duplicates(inplace=True)
data.duplicated().sum()
```

```python
#changing data types 
data.dtypes

data['id'] = data['id'].astype(object)
data['host_id'] = data['host_id'].astype(object)

data.dtypes
```

**Data Analysis**

Univariate Analysis

```python
#identifying outliers in price 

df = data[data['price']<1500] 

sns.boxplot(data=df,x='price')
```

```python
#Price distribution
data['price']
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df,x='price',bins=100)
plt.title("Price Distribution")
plt.ylabel("Frequency")
plt.show()
```

```python
# Availability of rooms
plt.figure(figsize=(6,5))
sns.histplot(data=df,x='availability_365')
plt.title("availability_365")
plt.ylabel("Frequency")
plt.show()
```

```python
data.dtypes
```

```python
df.groupby(by="neighbourhood_group")['price'].mean()
```

**Feature Engineering**

```python
#['price per bed']
df['price per bed'] = df['price']/df['beds']
df.head()
```

```python
#Price per bed 
df.groupby(by="neighbourhood_group")['price per bed'].mean()
```
